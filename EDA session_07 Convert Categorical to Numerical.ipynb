$MAP$

- Before applying map method first get the unique labels of the column
- for example case_status is a categorical column
- it has two unique labels are there
   - Denied
   - Certified
- Create a dictionary key as label , value as number
- d={'Certified':0,'Denied':1}
- this dictionary we need to map the case_status column

```python
path=r'C:\Users\reddy\Naresh IT\Data Science\python classes\EDA\visadataset.csv'
visa_df=pd.read_csv(path)
visa_df.head()
visa_df['case_status'].unique()
```

```python
d={'Certified':0,'Denied':1}
visa_df['case_status']=visa_df['case_status'].map(d)
```

$np.where$

- np.where requires 3 arguments
- conditions
- True
- False
- it applicable for only binary labels
- case status has only two  labels certified & Denied
- if case_status==certifird replace that as 0,other wise1

```python
path=r'C:\Users\reddy\Naresh IT\Data Science\python classes\EDA\visadataset.csv'
visa_df=pd.read_csv(path)
visa_df
con=visa_df['case_status']=='Certified'
visa_df['case_status']=np.where(con,0,1)
visa_df
```

***one hot Encoder***

- one hot encoder name says at a time one will on & other will off
- For Example case status has two labels
  - certified
  - Denied
- when you apply one hot encoding on case status, it creates two morer  extra columns
  - case_statua_certified
  - case-status_Denied


|case_status |case_status_certified  |case_status_denied|
|-|-|-|
|CERTIFIED|1|0|         
|DENIED|0|1|            

***ADVANTAGES***

- whem you develop ML model it is very important that the columns should be independent each other
- So here case-status creating two extra columns
- which are independent each other, which means the row values at a time only one column has 1
- columns are independent each other
- Which means 90 degress phase shift
- which means perpendicular each other
- which means othogonal each other

***DISADVANTAGES***

- The Disadvantage is if a column has  100 unique labels , 100 new columns can be created
- The data will become sparse, which means huge
- columns are more means, Dimensions are more
- The processing time   is more
- The memory consumption is more
- curse of Dimensionally
