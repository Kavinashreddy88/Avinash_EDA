$DROP$  $THE$  $COLUMNS$

- in order to drop the column
- we need use the drop method
- it takes 3 parameters
- drop column or row
- mention the column name
- axis 
- axis=1 reference as column
- axis-0 reference as row
- you want to create a new data frame or you want to over write existig data frame
- inplace=true

```python
city_names=['ank','ank','ank']
df['city']=city_names
df['NAME']=['sunnu','sai','jyo']
df['AGE']=['22','23','8']
df.drop('city',axis=1,inplace=True)
df
```
