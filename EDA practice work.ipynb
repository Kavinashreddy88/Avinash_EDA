BAR-CHART:

```python
path=r'C:\Users\reddy\Naresh IT\Data Science\python classes\EDA\visadataset.csv'
visa_df=pd.read_csv(path)
continent_vc=visa_df['education_of_employee'].value_counts()
l1=continent_vc.keys()
l2=continent_vc.values
continent_vc_df=pd.DataFrame(zip(l1,l2),columns=['education_of_employee','count'])
continent_vc_df
```

```python
plt.figure(figsize=(10,6)) 
plt.bar('education_of_employee','count',data=continent_vc_df)
plt.xlabel('education_of_employee')
plt.ylabel('count')
plt.title('Bar Chart of education_of_employee and their count')
plt.savefig('Datascience.jpg')
plt.show()
```
