```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path=r'C:\Users\reddy\Naresh IT\Data Science\python classes\EDA\visadataset.csv'
visa_df=pd.read_csv(path)
plt.figure(figsize=(10,6))
#l=['Asia', 'Europe', 'North America', 'South America', 'Africa','Oceania']
l=visa_df['continent'].value_counts().keys() # order provides automatically
sns.countplot(data=visa_df,x='continent',order=l)
plt.xlabel('continent')
plt.ylabel('count')
plt.title('Bar Chart')
plt.savefig('Datascience.jpg')
plt.show()
```
